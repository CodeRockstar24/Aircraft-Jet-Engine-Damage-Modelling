```python
# Check and Remove constant value columns from df001_train 
def drop_constant_value(dataframe):
    '''
    Function:
        - Deletes constant value columns in the data set.
        - A constant value is a value that is the same for all data in the data set.
        - A value is considered constant if the minimum (min) and maximum (max) values in the column are the same.
    Args:
        dataframe -> dataset to validate
    Returned value:
        dataframe -> dataset cleared of constant values
    '''

    # Creating a temporary variable to store a column name with a constant value
    constant_column = []

    # The process of finding a constant value by looking at the minimum and maximum values
    for col in dataframe.columns:
        min = dataframe[col].min()
        max = dataframe[col].max()

        # Append the column name if the min and max values are equal.
        if min == max:
            constant_column.append(col)

    # Delete column with constant value
    dataframe.drop(columns=constant_column, inplace=True)

    # return data
    return dataframe

# call function to drop constant value        
df001_train = drop_constant_value(df001_train)
```

```python
#check for outliers for each column 
def detect_outliers_iqr(df):
    outlier_sum = {}
    feature_cols = df.columns.drop(['cycle', 'RUL']) 
    
    for col in feature_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_sum[col] = len(outliers)
    
    return outlier_sum

# Check outliers in df001_train
outliers_fd001 = detect_outliers_iqr(df001_train)
print("Outliers in df001_train:")
for feature, count in outliers_fd001.items():
    print(f"{feature}: {count} outliers")
```

```python
# Generating Hazard curve RUL = 0 CONDITION 
# We can use the survival function to generate the hazard curve 
import lifelines
from lifelines import KaplanMeierFitter, NelsonAalenFitter
from lifelines.plotting import plot_lifetimes
# Step 1: Get RUL per engine
max_cycles = df_smoothing.groupby('engine_id')['cycle'].max().reset_index()
max_cycles.columns = ['engine_id', 'lifetime']
df_hazard = df_smoothing.merge(max_cycles, on='engine_id')
df_hazard['RUL'] = df_hazard['lifetime'] - df_hazard['cycle']

# Step 2: Survival Analysis 
# For each engine, one row at failure point
survival_data = df_hazard.groupby('engine_id').agg({'cycle': 'max'})
survival_data['event'] = 1  # all engines are failed in FD001
survival_data.reset_index(inplace=True)

# Step 3: Fit and Plot Hazard Curve
naf = NelsonAalenFitter()
naf.fit(survival_data['cycle'], event_observed=survival_data['event'])
plt.figure(figsize=(8,5))
naf.plot_hazard(bandwidth=10, color='blue', label='Hazard Rate')
plt.title('Hazard Curve for Jet Engines (FD001)')
plt.xlabel('Cycle')
plt.ylabel('Hazard Rate')
plt.grid(True)
plt.show()

```

```python
# Define health based on RUL
def assign_health(rul):
    if 0 <= rul <= 15:
        return 'High'
    elif 16 <= rul <= 30:
        return 'Medium'
    else:
        return 'Low'

# Prepare RUL and health index
max_cycles = df_smoothing.groupby('engine_id')['cycle'].max().reset_index()
max_cycles.columns = ['engine_id', 'lifetime']
df = df_smoothing.merge(max_cycles, on='engine_id')
df['RUL'] = df['lifetime'] - df['cycle']
df['health'] = df['RUL'].apply(assign_health)
df['event'] = 1  # all engines fail

# Plot hazard curves by health condition
naf = NelsonAalenFitter()
plt.figure(figsize=(10, 6))

for label in ['Low', 'Medium', 'High']:
    subset = df[df['health'] == label]
    naf.fit(subset['cycle'], event_observed=subset['event'], label=label)
    naf.plot_hazard(bandwidth=10)

plt.title('Hazard Rate by Health Index')
plt.xlabel('Cycle')
plt.ylabel('Hazard Rate')
plt.grid(True)
plt.tight_layout()
plt.legend(title='Health Index')
plt.show()
```

# WRAPPER METHOD (RFE (Recursive Feature elimination))

```python
from xgboost import XGBClassifier
from sklearn.feature_selection import RFE
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, roc_curve
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

X = df001_train.drop(columns=['failure', 'engine_id', 'RUL'])
y = df001_train['failure']


param_grid = {
    'clf__n_estimators': [100],
    'clf__max_depth': [3, 5, 7],
    'clf__learning_rate': [0.1]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

best_auc = 0
best_model = None
best_features = None
auc_scores = []

for k in range(5, X.shape[1] + 1):
    base_estimator = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
    rfe = RFE(estimator=base_estimator, n_features_to_select=k)
    X_rfe = rfe.fit_transform(X, y)
    selected_columns = X.columns[rfe.support_]

    pipeline = ImbPipeline([
        ('smote', SMOTE(random_state=42)),
        ('clf', XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42))
    ])

    grid = GridSearchCV(pipeline, param_grid, cv=cv, scoring='accuracy', n_jobs=-1)
    grid.fit(X_rfe, y)
    y_pred = grid.predict(X_rfe)
    y_pred_proba = grid.predict_proba(X_rfe)[:, 1]
    auc = roc_auc_score(y, y_pred_proba)

    if auc > best_auc:
        best_auc = auc
        best_model = grid.best_estimator_
        best_features = selected_columns

y_pred_final = best_model.predict(df001_train[best_features])
conf_matrix = confusion_matrix(y, y_pred_final)
acc = accuracy_score(y, y_pred_final)

print("Accuracy:", acc)
print("Top Features:", list(best_features))

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

from sklearn.metrics import classification_report
print(classification_report(y, y_pred_final))
```

```python
# IMPORT THE IMP MODULES
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import StackingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from imblearn.pipeline import Pipeline as ImbPipeline

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

base_learners = [
    ('rf', RandomForestClassifier(random_state=42)),
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)),
    ('lgbm', LGBMClassifier(random_state=42)),
    ('mlp', MLPClassifier(max_iter=1000, random_state=42)),
    ('dt', DecisionTreeClassifier(random_state=42))
]

meta_learner = LGBMClassifier(random_state=42)

stack = StackingClassifier(
    estimators=base_learners,
    final_estimator=meta_learner,
    cv=skf,
    n_jobs=-1
)

pipeline = ImbPipeline([
    ('SMOTE', SMOTE(random_state=42)),
    ('scaler', StandardScaler()),
    ('stack', stack)
])
 # Fit the model 
pipeline.fit(X_train, y_train)

scores = cross_val_score(pipeline, X_train, y_train, cv=skf, scoring='accuracy', n_jobs=-1)
print("Mean CV Accuracy:", scores.mean())
```
